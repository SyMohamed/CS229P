- **Softmax Function**

Instead of the sigmoid, multiclass logistic regression uses **Softmax**. It squashes a vector of raw scores (logits) into a probability distribution where the probabilities sum to **1**:

$$\text{Softmax}(z_i) = \frac{e^{z_i}}{\sum_{j=1}^{K} e^{z_j}}$$

- **Categorical Cross-Entropy Loss**

$$J(\theta) = -\frac{1}{m} \sum_{i=1}^{m} \sum_{k=1}^{K} y_{i,k} \log(\hat{y}_{i,k})$$

Where:
- $m$ = number of samples
- $K$ = number of classes
- $y_{i,k}$ = 1 if class $k$ is the correct label for sample $i$, else 0
- $\hat{y}_{i,k}$ = predicted probability for class $k$

> **Intuition:** the loss only cares about the probability assigned to the **correct** class. If the model assigns a low probability to the true class, the penalty becomes very large.
