```python
# Extract the features from the dataframe
price = df["price"].values
area = df["area"].values
beds = df["beds"].values
condo = df["condo"].values
location = df["location"].values

# Number of samples
N = len(price)

# Feature engineering
a = np.vstack(np.ones(N))  # Intercept term (ones)
b = np.vstack(area)  # Area
c = np.vstack([max(i - 1.5, 0) for i in np.hstack(area)])  # Max(area - 1.5, 0)
d = np.vstack(beds)  # Number of beds
e = np.vstack(condo)  # Condo indicator
# one hot vetor formed by f, g, h
f = np.vstack((location == 2).astype(int))  # Location == 2 indicator
g = np.vstack((location == 3).astype(int))  # Location == 3 indicator
h = np.vstack((location == 4).astype(int))  # Location == 4 indicator

# Combine all features into one matrix
X = np.hstack((a, b, c, d, e, f, g, h))
print(X.shape)
print(X[0,:])
```

## 3️⃣ Split the Data

> ✅ We split **before** any fitting to prevent information from the test set leaking into training.


```python
from sklearn.model_selection import train_test_split

# Step 4: Split the data into training and testing sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, price, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

## 4️⃣ Train the Model (three equivalent techniques)


### Method 1 — QR decomposition

```python
Q, R = np.linalg.qr(X_train)
theta_0 = np.linalg.solve(R, Q.T @ y_train)
theta_0
```

### Method 2 — Normal equation (matrix inversion)

```python
theta_1 = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train
theta_1
```

```python
np.allclose(theta_0, theta_1)
```

### Method 3 — `scikit-learn` `LinearRegression`

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=False)
model.fit(X_train, y_train)
theta_2 = model.coef_
theta_2, model.intercept_
```

## 5️⃣ Evaluate on the single split


```python
# Define the RMS function
def rms(errors):
    return np.sqrt(np.mean(np.square(errors)))


# Calculate the RMS error on the training set
rms_error_train = rms(np.matmul(X_train, theta_0) - y_train)

# Calculate the RMS error on the testing set
rms_error_test = rms(np.matmul(X_test, theta_0) - y_test)

# Output the results
print("Computed coefficients (theta):\n", theta_0)
print("\nRoot Mean Square Error (RMS) on Training Set:", rms_error_train)
print("Root Mean Square Error (RMS) on Testing Set:", rms_error_test)
```

> After evaluating the single-split model, we estimate generalisation more robustly using **5-fold cross-validation** and then average the coefficients across folds.

```python
# Assuming you have your predicted values for the test set
y_pred_train = np.matmul(X_train, theta_0)
y_pred_test = np.matmul(X_test, theta_0)

# Visualization: Actual vs Predicted
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_test, alpha=0.7, color='g', label='Predicted vs Actual')
plt.scatter(y_train, y_pred_train, alpha=0.7, color='b', label='Predicted vs Actual')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'r--', label='Perfect Prediction Line')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted House Prices')
plt.legend()
plt.show()
```

## 6️⃣ 5-Fold Cross-Validation


```python
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Store RMS errors for each fold
rms_errors_test = []
rms_errors_train = []
theta_list = []

# Perform cross-validation
for train_index, test_index in kf.split(X):
    # Split data into training and testing sets for the current fold
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = price[train_index], price[test_index]

    # Fit the model using the least squares method on the training data
    Q, R = np.linalg.qr(X_train)
    theta = np.linalg.solve(R, Q.T @ y_train)
    theta_list.append(theta)

    # Predict on the training set
    y_pred_train = np.matmul(X_train, theta)
    # Calculate RMS error for the current fold on the training set
    rms_error_train = rms(y_pred_train - y_train)
    rms_errors_train.append(rms_error_train)

    # Predict on the test set
    y_pred_test = np.matmul(X_test, theta)
    # Calculate RMS error for the current fold on the test set
    rms_error_test = rms(y_pred_test - y_test)
    rms_errors_test.append(rms_error_test)

# Calculate the average RMS error across all folds
average_rms_error_train = np.mean(rms_errors_train)
average_rms_error_test = np.mean(rms_errors_test)
# one may take the rms of the vec

# Output the results
print("RMS errors on the training set for each fold:", rms_errors_train)
print("RMS errors on the testing set for each fold:", rms_errors_test)
print("\nAverage RMS error on the training set across 5 folds:", average_rms_error_train)
print("Average RMS error on the testing set across 5 folds:", average_rms_error_test)
```

```python
# Calculate the average theta across all folds
average_theta = np.mean(theta_list, axis=0)

# Use the average theta to make predictions on the entire dataset
y_pred = np.matmul(X, average_theta)

# Visualization: Actual vs Predicted using the average theta
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_test, alpha=0.7, color='g', label='Predicted vs Actual')
plt.scatter(price, y_pred, alpha=0.7, color='b', label='Predicted vs Actual')
plt.plot([price.min(), price.max()], [price.min(), price.max()], 'r--', label='Perfect Prediction Line')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted House Prices Using Average Theta')
plt.legend()
plt.show()
```
